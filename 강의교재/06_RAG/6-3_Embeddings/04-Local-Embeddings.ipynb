## 1. Ollama Embeddings (권장)

Ollama(올라마)는 로컬 LLM 실행을 위한 가장 쉬운 도구예요. 모델 다운로드부터 실행까지 터미널 명령 두 개로 끝나요.

**설치 방법**:
1. [ollama.com](https://ollama.com)에서 OS별 설치 파일 다운로드
2. 터미널에서 임베딩 모델 다운로드: `ollama pull nomic-embed-text`
3. Ollama 서버가 자동으로 `http://localhost:11434`에서 실행돼요

> 🎯 **강의 포인트**: Ollama는 세 가지 로컬 임베딩 도구 중 **가장 먼저 시도해볼** 도구예요. 설치가 간단하고, LLM 추론 서버와 임베딩 서버를 하나로 관리할 수 있어서 RAG 스택 전체를 Ollama 하나로 구성할 수도 있어요.

> ⚠️ **자주 하는 실수**: Ollama가 실행 중이 아니면 오류가 발생해요. 터미널에서 `ollama serve` 명령으로 서버를 먼저 시작해야 해요.

````python

# ---------------------------------------------------
# Ollama 임베딩 모델 초기화
# ---------------------------------------------------

# ============================================================
# TODO: OllamaEmbeddings로 로컬 임베딩 모델을 초기화해보세요
# 힌트: OllamaEmbeddings(model="nomic-embed-text")
# 사전 조건: ollama pull nomic-embed-text 실행 필요
# 예상 결과: "Ollama Embeddings 초기화 완료" 출력
# ============================================================

from langchain_community.embeddings import OllamaEmbeddings
import numpy as np

# OllamaEmbeddings: 로컬 Ollama 서버를 통해 임베딩을 생성하는 클래스
# model: Ollama에서 다운로드한 임베딩 모델 이름
ollama_embeddings = OllamaEmbeddings(
    model="nomic-embed-text",  # 또는 "mxbai-embed-large", "chatfire/bge-m3:q8_0" 등
)

print("✅ Ollama Embeddings 초기화 완료")
print(f"  - 모델: nomic-embed-text")
print(f"  - 로컬 실행: http://localhost:11434")
````

````python

# ---------------------------------------------------
# Ollama로 임베딩 생성 및 유사도 검색 수행
# ---------------------------------------------------

# ============================================================
# TODO: Ollama로 쿼리와 문서를 임베딩하고 유사도를 계산해보세요
# 힌트: embed_query() → embed_documents() → 코사인 유사도 계산
# 예상 결과: 유사도 순으로 정렬된 문서 목록 (Ollama 미실행 시 오류 안내)
# ============================================================

from sklearn.metrics.pairwise import cosine_similarity

try:
    # 1단계: 쿼리 임베딩
    ollama_query_vec = ollama_embeddings.embed_query(query)
    
    # 2단계: 문서 임베딩
    ollama_doc_vecs = ollama_embeddings.embed_documents(texts)
    
    # 3단계: 코사인 유사도 계산
    # 내적(dot product)은 벡터 크기에 영향을 받아 값이 수백 단위로 나올 수 있음
    # 코사인 유사도는 방향만 비교하므로 0~1 사이의 직관적인 값을 반환
    ollama_similarities = cosine_similarity(
        [ollama_query_vec], ollama_doc_vecs
    )[0]
    ollama_sorted_indices = ollama_similarities.argsort()[::-1]
    
    print(f"\n[Ollama 검색 결과]")
    print(f"쿼리: {query}")
    print("=" * 60)
    for i, idx in enumerate(ollama_sorted_indices[:3], 1):
        print(f"[{i}위] 유사도: {ollama_similarities[idx]:.4f}")
        print(f"    {texts[idx][:50]}...")
        print()
        
except Exception as e:
    print(f"⚠️  Ollama 실행 오류: {e}")
    print("Ollama가 설치되어 있고 실행 중인지 확인하세요.")
    print("터미널에서 'ollama pull nomic-embed-text' 실행")
````

## 2. LlamaCpp Embeddings

LlamaCpp(라마씨피피)는 C++로 구현된 경량 LLM 실행 라이브러리예요. GGUF 형식의 모델 파일을 직접 로드해서 실행해요.

**설치 방법**:
```bash
pip install llama-cpp-python
```

**GGUF 모델 파일 준비**: [HuggingFace](https://huggingface.co/models?search=gguf+embedding)에서 GGUF 형식의 임베딩 모델을 다운로드해요.

> 💡 **실무 팁**: LlamaCpp는 모델 파일을 직접 관리할 수 있어서 특정 모델 버전을 고정해서 사용해야 하는 운영 환경에 적합해요. 또한 GGUF 형식의 양자화(Quantization) 모델을 사용하면 용량을 크게 줄일 수 있어요 (예: 7B 모델 ~4GB → ~2GB).

````python

# ---------------------------------------------------
# LlamaCpp 임베딩 모델 초기화 및 임베딩 생성
# ---------------------------------------------------

from huggingface_hub import hf_hub_download
from langchain_community.embeddings import LlamaCppEmbeddings

try:
    # 1단계: HuggingFace에서 GGUF 모델 다운로드 (~80MB, 최초 1회만 다운로드)
    # nomic-embed-text-v1.5: Ollama 섹션과 동일한 모델의 GGUF 버전
    # Q4_K_M: 4비트 양자화로 용량을 크게 줄인 버전 (원본 ~500MB → ~80MB)
    model_path = hf_hub_download(
        repo_id="nomic-ai/nomic-embed-text-v1.5-GGUF",
        filename="nomic-embed-text-v1.5.Q4_K_M.gguf",
    )
    print(f"✅ 모델 다운로드 완료: {model_path.split('/')[-1]}")

    # 2단계: LlamaCppEmbeddings 초기화
    # n_ctx=2048: 모델의 최대 컨텍스트 길이
    # n_batch=512: 한 번에 처리할 토큰 수
    llamacpp_embeddings = LlamaCppEmbeddings(
        model_path=model_path,
        n_ctx=2048,
        n_batch=512,
    )
    
    print("✅ LlamaCpp Embeddings 초기화 완료")
    print("  - 형식: GGUF (Q4_K_M 양자화)")
    print("  - 언어: C++ 기반")
    
    # 3단계: 임베딩 생성 및 유사도 검색
    llamacpp_query_vec = llamacpp_embeddings.embed_query(query)
    # embed_documents()는 내부에서 배치 처리를 시도하여 seq_id 충돌이 발생할 수 있음
    # 문서를 개별로 임베딩하면 이 문제를 회피할 수 있음
    llamacpp_doc_vecs = [llamacpp_embeddings.embed_query(t) for t in texts]
    
    print(f"  - 벡터 차원: {len(llamacpp_query_vec)}")

    llamacpp_similarities = cosine_similarity(
        [llamacpp_query_vec], llamacpp_doc_vecs
    )[0]
    llamacpp_sorted_indices = llamacpp_similarities.argsort()[::-1]

    print(f"\n[LlamaCpp 검색 결과]")
    print(f"쿼리: {query}")
    print("=" * 60)
    for i, idx in enumerate(llamacpp_sorted_indices[:3], 1):
        print(f"[{i}위] 유사도: {llamacpp_similarities[idx]:.4f}")
        print(f"    {texts[idx][:50]}...")
        print()

except Exception as e:
    print(f"⚠️  LlamaCpp 사용 불가: {e}")
    print("\n💡 LlamaCpp 사용 방법:")
    print("1. llama-cpp-python 설치: pip install llama-cpp-python")
    print("2. 위 코드가 HuggingFace에서 GGUF 모델을 자동 다운로드합니다")
````

## 3. GPT4All Embeddings

GPT4All(지피티포올)은 CPU에 최적화된 로컬 LLM 도구예요. 초기화 시 모델이 자동으로 다운로드되어 가장 간편하게 사용할 수 있어요.

**설치 방법**:
```bash
pip install gpt4all
```

> **주의**: 처음 실행 시 모델 파일이 자동 다운로드되어 수분이 걸릴 수 있어요. GPU가 없는 CPU 전용 환경에서 임베딩이 필요할 때 선택하세요.

````python

# ---------------------------------------------------
# GPT4All 임베딩 모델 초기화 및 유사도 검색
# ---------------------------------------------------

from langchain_community.embeddings import GPT4AllEmbeddings

try:
    # GPT4AllEmbeddings: CPU에 최적화된 로컬 임베딩 (모델 자동 다운로드)
    gpt4all_embeddings = GPT4AllEmbeddings()
    
    print("✅ GPT4All Embeddings 초기화 완료")
    print("  - CPU 최적화")
    print("  - 모델: 자동 다운로드")
    
    # 1단계: 쿼리 및 문서 임베딩
    gpt4all_query_vec = gpt4all_embeddings.embed_query(query)
    gpt4all_doc_vecs = gpt4all_embeddings.embed_documents(texts)
    
    # 2단계: 코사인 유사도 계산
    gpt4all_similarities = cosine_similarity(
        [gpt4all_query_vec], gpt4all_doc_vecs
    )[0]
    gpt4all_sorted_indices = gpt4all_similarities.argsort()[::-1]
    
    print(f"\n[GPT4All 검색 결과]")
    print(f"쿼리: {query}")
    print("=" * 60)
    for i, idx in enumerate(gpt4all_sorted_indices[:3], 1):
        print(f"[{i}위] 유사도: {gpt4all_similarities[idx]:.4f}")
        print(f"    {texts[idx][:50]}...")
        print()
        
except Exception as e:
    print(f"⚠️  GPT4All 사용 불가: {e}")
    print("\n💡 GPT4All 설치: pip install gpt4all")
    print("첫 실행 시 모델이 자동으로 다운로드됩니다.")
````
