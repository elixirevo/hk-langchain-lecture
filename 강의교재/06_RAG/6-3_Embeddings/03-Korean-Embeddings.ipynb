## Query vs Passage 비대칭 임베딩

일반적인 임베딩은 쿼리와 문서에 같은 모델을 사용해요. 하지만 **검색 최적화 임베딩**은 쿼리와 문서에 서로 다른 모델을 사용해서 검색 정확도를 높여요.

```mermaid

    QV["쿼리 벡터"]:::output

    D2["문서 2: 인사동 정보..."]:::storage

    SIM["유사도 계산<br/>(코사인 유사도)"]:::process
```

> 🔑 **핵심 개념**: **비대칭 임베딩(Asymmetric Embedding)**은 쿼리와 문서의 언어적 성격이 다르다는 점을 활용해요. 질문은 짧고 추상적이며, 문서는 길고 구체적이에요. 같은 모델로 두 가지를 처리하면 최적화가 어려워요.

## 1. Upstage Solar Embeddings 초기화

Query 모델과 Passage 모델을 각각 초기화해볼게요. API 키가 없으면 자동으로 HuggingFace 모델로 대체돼요.

````python

# ---------------------------------------------------
# Upstage Solar 임베딩 모델 초기화 (Query / Passage 분리)
# ---------------------------------------------------

# ============================================================
# TODO: Query 모델과 Passage 모델을 각각 초기화해보세요
# 힌트: UpstageEmbeddings(model="solar-embedding-1-large-query", api_key=...)
# 예상 결과: "Upstage Solar Embeddings 초기화 완료" 출력
# ============================================================

from langchain_upstage import UpstageEmbeddings
from langchain_huggingface.embeddings import HuggingFaceEmbeddings

if UPSTAGE_API_KEY:
    # 1단계: Query 전용 임베딩 모델 — 짧은 검색 질문에 최적화
    query_embeddings = UpstageEmbeddings(
        model="solar-embedding-1-large-query",
        api_key=UPSTAGE_API_KEY,
    )

    # 2단계: Passage 전용 임베딩 모델 — 긴 문서·답변 텍스트에 최적화
    passage_embeddings = UpstageEmbeddings(
        model="solar-embedding-1-large-passage",
        api_key=UPSTAGE_API_KEY,
    )

    print("✅ Upstage Solar Embeddings 초기화 완료")
    print("  - Query 모델: solar-embedding-1-large-query")
    print("  - Passage 모델: solar-embedding-1-large-passage")
else:
    print("⚠️ UPSTAGE_API_KEY가 없어 HuggingFace 임베딩으로 대체합니다.")
    print("   - intfloat/multilingual-e5-large-instruct 모델을 사용합니다.")
    fallback_model = "intfloat/multilingual-e5-large-instruct"
    query_embeddings = HuggingFaceEmbeddings(
        model_name=fallback_model,
        model_kwargs={"device": "cpu"},
        encode_kwargs={"normalize_embeddings": True},
    )
    passage_embeddings = query_embeddings
````

````python

# ---------------------------------------------------
# Query 모델로 사용자 검색 질문 임베딩
# ---------------------------------------------------

# ============================================================
# TODO: Query 모델을 사용해 검색 질문을 벡터로 변환해보세요
# 힌트: query_embeddings.embed_query(query_text)
# 예상 결과: 벡터 차원과 일부 값 출력
# ============================================================

# 사용자의 검색 질문
query = "서울에서 전통 문화를 체험할 수 있는 곳은?"

# 1단계: Query 모델로 질문 임베딩
# embed_query(): 검색 질문(짧은 텍스트)에 최적화된 메서드
embedded_query = query_embeddings.embed_query(query)

print(f"[쿼리] {query}")
print(f"벡터 차원: {len(embedded_query)}")
print(f"벡터 일부: {embedded_query[:5]}")
````

````python

# ---------------------------------------------------
# Passage 모델로 문서 배치 임베딩
# ---------------------------------------------------

# ============================================================
# TODO: Passage 모델을 사용해 문서 리스트를 벡터로 변환해보세요
# 힌트: passage_embeddings.embed_documents(texts)
# 예상 결과: 문서 수와 벡터 차원 출력
# ============================================================

# 1단계: Passage 모델로 문서 임베딩
# embed_documents(): 긴 문서(Passage)에 최적화된 배치 임베딩 메서드
embedded_documents = passage_embeddings.embed_documents(texts)

print(f"[문서 임베딩 완료]")
print(f"문서 수: {len(embedded_documents)}")
print(f"벡터 차원: {len(embedded_documents[0])}")
````

## 4. 한국어 검색 수행

Query 벡터와 Passage 벡터 간의 유사도를 계산해서 검색 결과를 얻어볼게요.

````python

# ---------------------------------------------------
# Query 벡터와 Passage 벡터 간 유사도 계산 및 검색 결과 출력
# ---------------------------------------------------

# ============================================================
# TODO: 벡터 내적으로 쿼리-문서 유사도를 계산하고 결과를 출력해보세요
# 힌트: np.array(embedded_query) @ np.array(embedded_documents).T
# 예상 결과: 인사동/경복궁 등 전통 문화 관련 문서가 상위 순위에 나와야 해요
# ============================================================

import numpy as np

# 1단계: 벡터 내적으로 유사도 계산
similarities = np.array(embedded_query) @ np.array(embedded_documents).T

# 2단계: 유사도 내림차순 정렬
sorted_indices = similarities.argsort()[::-1]

# 3단계: 검색 결과 출력
print(f"[쿼리] {query}")
print("=" * 80)
print("\n🔍 검색 결과:\n")
for i, idx in enumerate(sorted_indices, 1):
    print(f"[{i}위] 유사도: {similarities[idx]:.4f}")
    print(f"{texts[idx]}")
    print()
````

## 5. 다양한 쿼리 테스트

검색 함수를 만들어서 여러 가지 한국어 질문으로 성능을 확인해볼게요.

> 🎯 **강의 포인트**: 같은 벡터 DB라도 어떤 쿼리를 넣느냐에 따라 검색 결과가 크게 달라져요. "해수욕장"과 "바다를 볼 수 있는 곳"은 다른 단어지만 의미상 같은 문서를 찾아야 해요. 임베딩 기반 검색이 키워드 검색보다 강력한 이유가 여기 있어요.

````python

# ---------------------------------------------------
# 한국어 검색 함수 정의 및 다양한 쿼리 테스트
# ---------------------------------------------------

# ============================================================
# TODO: 아래 검색 함수를 완성하고 다양한 쿼리로 테스트해보세요
# 힌트: embed_query() → 내적 유사도 → argsort()[::-1][:top_k]
# 예상 결과: 각 쿼리에 맞는 관광지가 상위에 검색되어야 해요
# ============================================================

def search_korean_documents(query_text, top_k=3):
    """한국어 쿼리로 문서 검색"""
    # 1단계: Query 모델로 질문 임베딩
    query_vec = query_embeddings.embed_query(query_text)
    
    # 2단계: 유사도 계산
    sims = np.array(query_vec) @ np.array(embedded_documents).T
    
    # 3단계: Top-K 결과 반환
    top_indices = sims.argsort()[::-1][:top_k]
    
    return [(idx, sims[idx]) for idx in top_indices]

# 다양한 쿼리로 검색 테스트
test_queries = [
    "바다를 볼 수 있는 관광지는?",
    "야경이 아름다운 곳 추천해줘",
    "역사적인 건축물을 보고 싶어요",
]

for query_text in test_queries:
    print(f"\n🔍 쿼리: '{query_text}'")
    print("-" * 80)
    
    results = search_korean_documents(query_text, top_k=2)
    
    for rank, (idx, score) in enumerate(results, 1):
        print(f"  [{rank}위] 유사도: {score:.4f}")
        print(f"  {texts[idx][:60]}...")
        print()
````

### 핵심 요약

| 구분 | 사용 모델 | 메서드 | 시점 |
|------|----------|--------|------|
| 사용자 질문 | Query 모델 | `embed_query()` | 매 검색마다 |
| 문서·답변 | Passage 모델 | `embed_documents()` | 색인 시 1회 |

**비대칭 임베딩의 핵심**: 쿼리와 문서의 텍스트 성격이 다르기 때문에 각각에 최적화된 모델을 쓰면 검색 정확도가 높아져요.

> 🔑 **핵심 개념**: 비대칭 임베딩은 한국어 RAG에서 특히 효과적이에요. 한국어는 조사와 어미 변화가 많아서 동일한 의미도 다양한 형태로 표현되는데, 이런 구조적 특성을 이해한 모델이 더 좋은 검색 결과를 만들어요.
