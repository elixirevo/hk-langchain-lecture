PDF, Word, PowerPoint 등 다양한 형식을 지원하며, 표와 이미지를 고품질로 추출합니다.

- `result_type="markdown"` 옵션으로 구조화된 출력을 얻어요
- 다른 PDF 로더들과 성능·비용을 비교해서 선택 기준을 파악해요

## 사전 지식
- 12-UpstageLayoutAnalysisLoader 완료 (유/무료 AI 파서 차이 이해)
- LlamaCloud API 키 발급: [cloud.llamaindex.ai](https://cloud.llamaindex.ai/)
- `pip install llama-parse langchain-community` 설치

---

> **RAG 파이프라인 위치**: Load 단계 — Document Loader 시리즈의 마지막이에요. AI 기반 파서 중 **최고 정밀도**를 자랑하는 LlamaParse를 다뤄요.


> 🎯 **강의 포인트**: LlamaParse의 결정적인 차별점은 **이미지 설명 생성**입니다. 다른 로더들이 이미지를 무시하는 반면, LlamaParse는 이미지 내용을 텍스트로 변환합니다. 차트나 다이어그램이 중요한 문서에서 압도적입니다.

```python
from dotenv import load_dotenv
import os
import nest_asyncio

load_dotenv()
nest_asyncio.apply()

# LlamaCloud API 키 필요
LLAMA_CLOUD_API_KEY = os.getenv("LLAMA_CLOUD_API_KEY")
FILE_PATH = "./data/sample-rag-brief.pdf"
```

> 🔑 **핵심 개념**: `result_type="markdown"` 옵션이 핵심입니다. 표는 Markdown 표 형식으로, 헤더는 `#` 기호로 변환되어 구조화된 출력이 나옵니다. 이후 `MarkdownHeaderTextSplitter`와 결합하면 완벽한 파이프라인을 구성할 수 있습니다.

## LlamaParse 특징

### 주요 기능
- **AI 기반 파싱**: GPT-4V 수준의 정확도
- **복잡한 표 인식**: 병합된 셀, 중첩 표 등 처리
- **이미지 설명**: 이미지 내용을 텍스트로 변환
- **다양한 형식**: PDF, DOCX, PPTX 지원
- **Markdown 출력**: 구조화된 Markdown 형식 제공

### 설정
**필요한 패키지**: `pip install llama-parse langchain-community`

**API 키 발급**: [LlamaCloud](https://cloud.llamaindex.ai/)


> 💡 **실무 팁**: LlamaParse는 유료 서비스지만 무료 티어로도 1,000페이지/일 처리가 가능합니다. 고품질이 필요한 핵심 문서에 선별적으로 사용하고, 일반 문서는 무료 로더를 사용하는 하이브리드 전략이 비용 효율적입니다.

```python
# ============================================================
# TODO: LlamaParse로 PDF를 Markdown 형식으로 파싱해보세요
# 힌트: LlamaParse(api_key=LLAMA_CLOUD_API_KEY, result_type="markdown") → loader.load_data(FILE_PATH)
# 예상 결과: PDF가 Markdown 형식으로 변환되어 출력됩니다
# ============================================================

from llama_parse import LlamaParse

if LLAMA_CLOUD_API_KEY:
    # 1단계: LlamaParse 생성
    # - result_type="markdown": 구조화된 Markdown 형식으로 출력 (표, 헤더 보존)
    # - verbose=True: API 처리 진행 상황 출력
    loader = LlamaParse(
        api_key=LLAMA_CLOUD_API_KEY,
        result_type="markdown",  # "markdown" 또는 "text"
        verbose=True
    )
    
    # 2단계: 문서 파싱 (AI API 호출)
    docs = loader.load_data(FILE_PATH)
    
    print(f"추출된 문서 수: {len(docs)}")
    print(f"\nMarkdown 출력 미리보기:")
    print(docs[0].text[:500])
else:
    print("⚠️ LLAMA_CLOUD_API_KEY가 설정되지 않았습니다.")
    print("LlamaCloud에서 API 키를 발급받아 .env 파일에 추가하세요.")
    print("\n💡 API 키 발급:")
    print("1. https://cloud.llamaindex.ai/ 방문")
    print("2. 회원가입 후 API Keys 메뉴에서 생성")
    print("3. .env 파일에 LLAMA_CLOUD_API_KEY=your-key 추가")
```
