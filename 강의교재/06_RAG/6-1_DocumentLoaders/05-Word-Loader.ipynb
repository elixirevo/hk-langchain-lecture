# Word 문서 로더

Microsoft Word(`.docx`, `.doc`)는 전 세계적으로 가장 많이 사용되는 워드프로세서 형식입니다.

이 노트북에서는 Word 문서를 LangChain Document로 변환하는 방법을 학습합니다.

## 주요 로더

1. **Docx2txtLoader**: 빠르고 간단한 텍스트 추출
2. **UnstructuredWordDocumentLoader**: 문서 구조 보존 및 상세 분석


## 환경 설정


````python
from dotenv import load_dotenv
import os

load_dotenv()

# 샘플 Word 파일 경로
FILE_PATH = "./data/sample-word-document.docx"

# 데이터 파일 경로 확인
if not os.path.exists(FILE_PATH):
    raise FileNotFoundError(f"파일을 찾을 수 없습니다: {FILE_PATH}\n현재 디렉토리: {os.getcwd()}")
````

## 1. Docx2txtLoader

`Docx2txtLoader`는 `.docx` 파일을 빠르게 텍스트로 변환하는 로더입니다.

### 특징

- 빠른 텍스트 추출
- 간단한 사용법
- 순수 텍스트 위주

### 필요한 패키지

```bash
pip install docx2txt
```


> ⚠️ **자주 하는 실수**: `pip install docx`가 아니라 반드시 `pip install python-docx`를 설치해야 합니다. 구버전 `docx` 패키지는 `No module named 'exceptions'` 오류를 일으킵니다.

````python
from langchain_community.document_loaders import Docx2txtLoader

# ============================================================
# TODO: Docx2txtLoader로 Word 문서를 로딩해보세요
# 힌트: Docx2txtLoader(FILE_PATH)로 초기화 후 load()를 호출합니다
# 예상 결과: 문서 개수 1개와 Document 타입이 출력됩니다
# ============================================================

# 1단계: Docx2txtLoader 생성
# - Docx2txtLoader: .docx 파일을 빠르게 순수 텍스트로 변환하는 로더
loader = Docx2txtLoader(FILE_PATH)

# 2단계: 문서 로드
docs = loader.load()

print(f"로드된 문서 개수: {len(docs)}")
print(f"\n첫 번째 문서의 타입: {type(docs[0])}")
````

### 문서 내용 확인


````python
# 첫 번째 문서의 내용 출력 (처음 300자)
print("=" * 60)
print("📄 문서 내용 미리보기:")
print("=" * 60)
print(docs[0].page_content[:300])

print("\n" + "=" * 60)
print("🏷️  메타데이터:")
print("=" * 60)
print(docs[0].metadata)
````

> 🎯 **강의 포인트**: `mode="elements"`를 사용하면 각 요소의 `metadata["category"]`로 `Title`, `NarrativeText` 등을 구분할 수 있습니다. 제목만 뽑아 목차를 구성하거나, 본문 단락만 벡터화하는 등 세밀한 제어가 가능합니다.

## 2. UnstructuredWordDocumentLoader

`UnstructuredWordDocumentLoader`는 문서의 구조 정보를 보존하면서 로드합니다.

### 특징

- 문서 구조 보존 (제목, 단락 등)
- `mode="elements"`로 요소별 분리 가능
- 더 상세한 메타데이터 제공

### 필요한 패키지

```bash
pip install unstructured python-docx
```

**⚠️ 주의**: `docx` 패키지가 아닌 `python-docx` 패키지를 설치해야 합니다!


````python
from langchain_community.document_loaders import UnstructuredWordDocumentLoader

# ============================================================
# TODO: UnstructuredWordDocumentLoader로 Word 문서를 로딩해보세요
# 힌트: UnstructuredWordDocumentLoader(FILE_PATH)로 초기화합니다
# 예상 결과: 문서 개수와 내용 200자가 출력됩니다
# ============================================================

# 1단계: UnstructuredWordDocumentLoader 생성 (기본 모드)
# - 기본 모드: 모든 요소를 하나의 Document로 결합
# - Docx2txtLoader보다 느리지만 구조 정보(메타데이터)를 더 많이 제공
loader = UnstructuredWordDocumentLoader(FILE_PATH)

# 2단계: 문서 로드
docs = loader.load()

print(f"로드된 문서 개수: {len(docs)}")
print(f"\n첫 번째 문서의 내용 (처음 200자):")
print(docs[0].page_content[:200])
````

### 메타데이터 확인


````python
# 메타데이터 출력
print("=" * 60)
print("🔍 메타데이터:")
print("=" * 60)
print(docs[0].metadata)
````

### Elements 모드

`mode="elements"`를 사용하면 문서를 요소별로 분리할 수 있습니다. 각 제목, 단락 등이 별도의 Document로 생성됩니다.


> 💡 **실무 팁**: 텍스트만 빠르게 추출할 때는 `Docx2txtLoader`, 문서 구조(제목/단락 계층)가 중요하면 `UnstructuredWordDocumentLoader(mode="elements")`를 선택하세요. 두 가지를 모두 실행해서 결과를 비교해보는 것이 좋습니다.

````python
# ============================================================
# TODO: mode="elements"로 Word 문서를 요소별로 분리해보세요
# 힌트: UnstructuredWordDocumentLoader(FILE_PATH, mode="elements")
# 예상 결과: 128개의 요소가 생성되고 각 요소의 메타데이터에 category가 포함됩니다
# ============================================================

# 1단계: elements 모드로 로더 생성
# - mode="elements": 제목, 단락, 표 등 각 요소를 별도 Document로 분리
# - 각 Document의 metadata에 category 정보 저장 (Title, NarrativeText 등)
loader = UnstructuredWordDocumentLoader(FILE_PATH, mode="elements")

# 2단계: 문서 로드
docs = loader.load()

print(f"Elements 모드 문서 개수: {len(docs)}")
print(f"\n처음 3개 요소:")
for i, doc in enumerate(docs[:3]):
    print(f"\n[요소 {i}]")
    print(f"내용: {doc.page_content[:100]}...")
    print(f"메타데이터: {doc.metadata}")
````

## 💡 핵심 정리

### Docx2txtLoader

```python
from langchain_community.document_loaders import Docx2txtLoader

loader = Docx2txtLoader("file.docx")
docs = loader.load()
```

- ✅ **장점**: 빠른 속도, 간단한 사용법
- ✅ **추천**: 순수 텍스트만 필요할 때
- ⚠️ **한계**: 문서 구조 정보 손실

### UnstructuredWordDocumentLoader

```python
from langchain_community.document_loaders import UnstructuredWordDocumentLoader

# 기본 모드 (전체 문서를 하나의 Document로)
loader = UnstructuredWordDocumentLoader("file.docx")
docs = loader.load()

# Elements 모드 (요소별로 분리)
loader = UnstructuredWordDocumentLoader("file.docx", mode="elements")
docs = loader.load()
```

- ✅ **장점**: 문서 구조 보존, 상세한 메타데이터
- ✅ **추천**: 제목/단락 구분이 중요할 때
- ⚠️ **주의**: `python-docx` 패키지 필요 (오래된 `docx`는 오류 발생)

### 선택 가이드

| 상황 | 추천 로더 |
|------|----------|
| 빠른 텍스트 추출 | **Docx2txtLoader** |
| 문서 구조 보존 필요 | UnstructuredWordDocumentLoader |
| 제목/단락 분리 필요 | UnstructuredWordDocumentLoader (elements) |
| 간단한 사용 | **Docx2txtLoader** |

### 패키지 설치 주의사항

pip install python-docx

# ❌ 오래된 패키지 (오류 발생)
pip install docx  # ModuleNotFoundError: No module named 'exceptions'
