## Query vs Passage 비대칭 임베딩

## 4. 한국어 검색 수행

Query 벡터와 Passage 벡터 간의 유사도를 계산해서 검색 결과를 얻어볼게요.

````python
# ---------------------------------------------------
# Query 벡터와 Passage 벡터 간 유사도 계산 및 검색 결과 출력
# ---------------------------------------------------

# ============================================================
# TODO: 벡터 내적으로 쿼리-문서 유사도를 계산하고 결과를 출력해보세요
# 힌트: np.array(embedded_query) @ np.array(embedded_documents).T
# 예상 결과: 인사동/경복궁 등 전통 문화 관련 문서가 상위 순위에 나와야 해요
# ============================================================

import numpy as np

# 1단계: 벡터 내적으로 유사도 계산
similarities = np.array(embedded_query) @ np.array(embedded_documents).T

# 2단계: 유사도 내림차순 정렬
sorted_indices = similarities.argsort()[::-1]

# 3단계: 검색 결과 출력
print(f"[쿼리] {query}")
print("=" * 80)
print("\n:mag: 검색 결과:\n")
for i, idx in enumerate(sorted_indices, 1):
    print(f"[{i}위] 유사도: {similarities[idx]:.4f}")
    print(f"{texts[idx]}")
    print()
````

## 5. 다양한 쿼리 테스트

검색 함수를 만들어서 여러 가지 한국어 질문으로 성능을 확인해볼게요.

````python

# ---------------------------------------------------
# 한국어 검색 함수 정의 및 다양한 쿼리 테스트
# ---------------------------------------------------

# ============================================================
# TODO: 아래 검색 함수를 완성하고 다양한 쿼리로 테스트해보세요
# 힌트: embed_query() → 내적 유사도 → argsort()[::-1][:top_k]
# 예상 결과: 각 쿼리에 맞는 관광지가 상위에 검색되어야 해요
# ============================================================

def search_korean_documents(query_text, top_k=3):
    """한국어 쿼리로 문서 검색"""
    # 1단계: Query 모델로 질문 임베딩
    query_vec = query_embeddings.embed_query(query_text)
    
    # 2단계: 유사도 계산
    sims = np.array(query_vec) @ np.array(embedded_documents).T
    
    # 3단계: Top-K 결과 반환
    top_indices = sims.argsort()[::-1][:top_k]
    
    return [(idx, sims[idx]) for idx in top_indices]

# 다양한 쿼리로 검색 테스트
````
