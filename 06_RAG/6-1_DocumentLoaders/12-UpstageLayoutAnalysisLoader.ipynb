## 핵심 정리 및 마무리

### 카테고리 종류와 활용

| 카테고리 | 설명 | 활용 |
|---------|------|------|
| `heading` | 제목/헤더 | 문서 구조 파악 |
| `paragraph` | 일반 단락 | 본문 내용 |
| `table` | 표 데이터 | 구조화된 데이터 추출 |
| `figure` | 그림/이미지 | 시각 자료 인식 |

### 로더 선택 기준

| 상황 | 추천 로더 |
|------|---------|
| 간단한 텍스트 PDF | `PyPDFLoader` (무료, 빠름) |
| 표가 많은 문서 | `PDFPlumberLoader` (무료, 중간 정확도) |
| 한글 문서 + 복잡한 레이아웃 | `UpstageLayoutAnalysisLoader` |
| 최고 정확도 필요 | `LlamaParse` |

### 실무 팁
> 재무제표, 연구 보고서처럼 **표가 많고 레이아웃이 복잡한 문서**에서 Upstage의 강점이 드러나요. 특히 한글 문서 인식 성능이 탁월합니다.
