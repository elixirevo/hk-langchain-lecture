## 2. UnstructuredWordDocumentLoader

`UnstructuredWordDocumentLoader`는 문서의 구조 정보를 보존하면서 로드합니다.

### 특징

- 문서 구조 보존 (제목, 단락 등)
- `mode="elements"`로 요소별 분리 가능
- 더 상세한 메타데이터 제공

### 필요한 패키지

```bash
pip install unstructured python-docx
```

**⚠️ 주의**: `docx` 패키지가 아닌 `python-docx` 패키지를 설치해야 합니다!

````python
from langchain_community.document_loaders import UnstructuredWordDocumentLoader

# ============================================================
# TODO: UnstructuredWordDocumentLoader로 Word 문서를 로딩해보세요
# 힌트: UnstructuredWordDocumentLoader(FILE_PATH)로 초기화합니다
# 예상 결과: 문서 개수와 내용 200자가 출력됩니다
# ============================================================

# 1단계: UnstructuredWordDocumentLoader 생성 (기본 모드)
# - 기본 모드: 모든 요소를 하나의 Document로 결합
# - Docx2txtLoader보다 느리지만 구조 정보(메타데이터)를 더 많이 제공
loader = UnstructuredWordDocumentLoader(FILE_PATH)
# 2단계: 문서 로드
docs = loader.load()
````

### Elements 모드

`mode="elements"`를 사용하면 문서를 요소별로 분리할 수 있습니다. 각 제목, 단락 등이 별도의 Document로 생성됩니다.

> **실무 팁**: 텍스트만 빠르게 추출할 때는 `Docx2txtLoader`, 문서 구조(제목/단락 계층)가 중요하면 `UnstructuredWordDocumentLoader(mode="elements")`를 선택하세요. 두 가지를 모두 실행해서 결과를 비교해보는 것이 좋습니다.

````python
# ============================================================
# TODO: mode="elements"로 Word 문서를 요소별로 분리해보세요
# 힌트: UnstructuredWordDocumentLoader(FILE_PATH, mode="elements")
# 예상 결과: 128개의 요소가 생성되고 각 요소의 메타데이터에 category가 포함됩니다
# ============================================================

# 1단계: elements 모드로 로더 생성
# - mode="elements": 제목, 단락, 표 등 각 요소를 별도 Document로 분리
# - 각 Document의 metadata에 category 정보 저장 (Title, NarrativeText 등)
loader = UnstructuredWordDocumentLoader(FILE_PATH, mode="elements")
# 2단계: 문서 로드
docs = loader.load()

print(docs[1].page_content[:100])
print(docs[0].metadata)
````

## 💡 핵심 정리

### Docx2txtLoader

```python
from langchain_community.document_loaders import Docx2txtLoader

loader = Docx2txtLoader("file.docx")
docs = loader.load()
```

- ✅ **장점**: 빠른 속도, 간단한 사용법
- ✅ **추천**: 순수 텍스트만 필요할 때
- ⚠️ **한계**: 문서 구조 정보 손실

### UnstructuredWordDocumentLoader

```python
from langchain_community.document_loaders import UnstructuredWordDocumentLoader

# 기본 모드 (전체 문서를 하나의 Document로)
loader = UnstructuredWordDocumentLoader("file.docx")
docs = loader.load()

# Elements 모드 (요소별로 분리)
loader = UnstructuredWordDocumentLoader("file.docx", mode="elements")
docs = loader.load()
```

- ✅ **장점**: 문서 구조 보존, 상세한 메타데이터
- ✅ **추천**: 제목/단락 구분이 중요할 때
- ⚠️ **주의**: `python-docx` 패키지 필요 (오래된 `docx`는 오류 발생)

### 선택 가이드

| 상황 | 추천 로더 |
|------|----------|
| 빠른 텍스트 추출 | **Docx2txtLoader** |
| 문서 구조 보존 필요 | UnstructuredWordDocumentLoader |
| 제목/단락 분리 필요 | UnstructuredWordDocumentLoader (elements) |
| 간단한 사용 | **Docx2txtLoader** |
