## 핵심 정리 및 마무리

### glob 패턴 가이드

| 패턴 | 설명 | 예시 결과 |
|------|------|-----------|
| `*.txt` | 현재 폴더의 txt 파일만 | `doc.txt` 포함, `sub/doc.txt` 제외 |
| `**/*.txt` | 모든 하위 폴더의 txt 파일 | `doc.txt`와 `sub/doc.txt` 모두 포함 |
| `**/chapter*.md` | chapter로 시작하는 md 파일 | `chapter1.md` 포함, `intro.md` 제외 |

### 실무 팁
> 대량 파일을 처리할 때는 `silent_errors=True`를 사용해서 일부 파일 오류로 전체 작업이 중단되는 것을 막아요. `show_progress=True`로 진행 상황을 모니터링할 수 있어요.
